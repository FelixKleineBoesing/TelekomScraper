```python
import sys
from pathlib import Path
sys.path.append(str(Path("../").absolute()))
```

```python
import pandas as pd
data = pd.read_json("../data/cleaned_html.json")
```

```python
data["texts"] = data.texts.apply(lambda x: [t for t in x if t != "expand_outline-optimized"])
data["concat_texts"] = data["texts"].str.join(" ").replace("  ", " ")
```

```python
data["length_texts"] = data["concat_texts"].str.len()
```

```python
import plotly.express as px
fig = px.histogram(data, x="length_texts")
fig.show()
```
